```r
# import list
not_there <- read.table('not_there.txt', header=FALSE)
dim(not_there)[1]
not_there[1:5,1]
```

```r
# look up all names 
now_there <- data.frame(Entrez_ID=character(), Symbol=character(), stringsAsFactors=FALSE)

for (i in 1:dim(not_there)[1]){
    r_search <- entrez_summary(db='gene', id=not_there[i,1])
    r_search <- r_search$name
    
    now_there[i,1] <- not_there[i,1]
    now_there[i,2] <- r_search
    r_search <- ''
}
```

```r
write.table(now_there, file = "now_there.csv", row.names=FALSE, col.names=FALSE, na="")
```
